### Day 1

```python
with open('hyperskill-dataset-118978279.txt', 'r') as file:
    errors = file.readlines()

selected_errors = []
for line in errors:
	parts = line.split()
	error_time = parts[0]
	error_name = parts[1]
	if error_time >= '15:00' and error_time <= '15:30':
		selected_errors += [line]

error_frequencies = {}
for line in errors:
	parts = line.split()
	error_time = parts[0]
	error_name = parts[1]
	if error_name in error_frequencies:
		error_frequencies[error_name] += 1
	else:
		error_frequencies[error_name] = 1

selected_frequencies = {}
for line in selected_errors:
	parts = line.split()
	error_name = parts[1]
	if error_name in selected_frequencies:
		selected_frequencies[error_name] += 1
	else:
		selected_frequencies[error_name] = 1

print("All errors:")
for error_name, frequency in error_frequencies.items():
	print(f"{error_name}: {frequency}")

print()
print("Selected errors:")
for error_name, frequency in selected_frequencies.items():
	print(f"{error_name}: {frequency}")
```

```python
with open('hyperskill-dataset-119000311.txt', 'r') as file:
    sweetness_levels = file.readlines()

target_sweetness = int(sweetness_levels[0].strip())
cocoa_sweetnesses = sweetness_levels[1].strip().split(',')
cocoa_sweetnesses = [int(level) for level in cocoa_sweetnesses]

error = float('inf')
sweetness_result = None

for i in range(len(cocoa_sweetnesses)):
	for j in range(i + 1, len(cocoa_sweetnesses)):
		sweetness = (cocoa_sweetnesses[i] + cocoa_sweetnesses[j]) // 2
		if abs(sweetness - target_sweetness) < error:
			error = abs(sweetness - target_sweetness)
			sweetness_result = sweetness

# round up to the nearest integer
from math import ceil
print(ceil(sweetness_result))
```

### Day 3

```python
with open('hyperskill-dataset-119010168.txt', 'r') as file:
    passwords = file.readlines()

passwords = [line.strip() for line in passwords]

scores = {}
for password in passwords:
	score = len(password)

	if not any(char.islower() for char in password):
		score *= 0.75
	if not any(char.isupper() for char in password):
		score *= 0.75
	if not any(char.isdigit() for char in password):
		score *= 0.75
	if not any(char in '!@#$%^&*' for char in password):
		score *= 0.75

	char_counts = {}
	for char in password:
		if char in char_counts:
			char_counts[char] += 1
		else:
			char_counts[char] = 1
	max_occurrence = max(char_counts.values())
	if max_occurrence / len(password) > 0.3:
		score -= max_occurrence

	scores[password] = score

print("Password scores:")
for password, score in scores.items():
	print(f"{password}: {score}")

max_score_passwords = [pwd for pwd, sc in scores.items() if sc == max(scores.values())]
print("Password(s) with the highest score:")
for password in max_score_passwords:
	print(password)
```

### Day 4

```python
with open('hyperskill-dataset-119028174.txt', 'r') as file:
    actions = file.readlines()

log = []
for action in actions:
    line = action.strip().split(',')
    log.append((line[0], line[1], int(line[2])))

forks = {}
contentions = 0
for entry in log:
	if entry[1] == 'pick':
		if entry[2] not in forks:
			forks[entry[2]] = entry[0]
		elif forks[entry[2]] != entry[0]:
			contentions += 1
	elif entry[1] == 'release':
		if entry[2] in forks and forks[entry[2]] == entry[0]:
			del forks[entry[2]]
print(contentions)
```

### Day 6

```python
from collections import deque, defaultdict

with open('hyperskill-dataset-119096205.txt', 'r') as file:
    lines = file.readlines()

start = lines[0].strip()
connections = [tuple(line.strip().split(',')) for line in lines[1:]]

graph = defaultdict(list)
for a, b in connections:
	graph[a].append(b)
	graph[b].append(a)
visited = {start: 0}
queue = deque([start])

while queue:
	node = queue.popleft()
	for neighbor in graph[node]:
		if neighbor not in visited:
			visited[neighbor] = visited[node] + 1
			queue.append(neighbor)

max_distance = max(visited.values())
farthest_nodes = [node for node, distance in visited.items() if distance == max_distance]
print(", ".join(sorted(farthest_nodes)))
```

### Day 9

```python
import heapq

with open('hyperskill-dataset-119144514.txt', 'r') as file:
    lines = file.readlines()

hazard_levels = [line.strip().split(',') for line in lines]
hazard_levels = [[int(level) for level in row] for row in hazard_levels]
print(hazard_levels)
hazard_map = {}
for i in range(len(hazard_levels)):
	for j in range(len(hazard_levels[0])):
		hazard_map[(i, j)] = hazard_levels[i][j]
print(hazard_map)

start = (0, 0)
end = (len(hazard_levels) - 1, len(hazard_levels[0]) - 1)
heap = [(hazard_map[start], start)]
visited = set()
min_hazard = {start: hazard_map[start]}
while heap:
	current_hazard, current_pos = heapq.heappop(heap)
	if current_pos in visited:
		continue
	visited.add(current_pos)
	if current_pos == end:
		break
	x, y = current_pos
	for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
		neighbor = (x + dx, y + dy)
		if 0 <= neighbor[0] < len(hazard_levels) and 0 <= neighbor[1] < len(hazard_levels[0]):
			new_hazard = current_hazard + hazard_map[neighbor]
			if neighbor not in min_hazard or new_hazard < min_hazard[neighbor]:
				min_hazard[neighbor] = new_hazard
				heapq.heappush(heap, (new_hazard, neighbor))
print(min_hazard[end])
```

### Day 10

```python
with open('hyperskill-dataset-119184299.txt', 'r') as file:
    lines = file.readlines()

sequences = [line.strip().split(',') for line in lines]
sequences = [[int(num) for num in seq] for seq in sequences]

def find_longest_common_subarray(seqs):
    if not seqs:
        return []
    
    first_seq = seqs[0]
    max_length = 0
    best_subseq = []
    
    for start in range(len(first_seq)):
        for end in range(start + 1, len(first_seq) + 1):
            candidate = first_seq[start:end]
            
            found_in_all = True
            for seq in seqs[1:]:
                found = False
                for i in range(len(seq) - len(candidate) + 1):
                    if seq[i:i+len(candidate)] == candidate:
                        found = True
                        break
                if not found:
                    found_in_all = False
                    break
            
            if found_in_all and len(candidate) > max_length:
                max_length = len(candidate)
                best_subseq = candidate
    
    return best_subseq

longest_subseq = find_longest_common_subarray(sequences)
print(len(longest_subseq))
print(",".join(str(num) for num in longest_subseq))
```
